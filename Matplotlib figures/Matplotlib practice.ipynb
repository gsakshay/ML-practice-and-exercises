```python
heart_dataset = pd.read_csv("heart-disease.csv")
heart_dataset.head()
```

```python
heart_dataset["age"].plot.hist();
```

```python
heart_dataset.plot.hist(figsize = (10, 20), subplots = True);
```

```python
over_60 = heart_dataset[heart_dataset["age"] > 60]
over_60.size
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
over_60.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target"
            ,ax = ax);
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax.scatter(x = over_60["age"],
           y = over_60["chol"],
           c = over_60["target"])
ax.set(title = "heart-functionality and cholesterol rate of patients over age 60", xlabel = "age", ylabel = "cholesterol")
ax.legend(*scatter.legend_elements(), title = "target")
ax.axhline(over_60["chol"].mean(), linestyle = "--");
```

```python
over_60.head()
```

#### subplot of chol, age , thalach

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1,  sharex = True, figsize = (10,10) )


ax0.scatter(x = over_60["age"],
              y = over_60["chol"],
           c = over_60["target"])
ax0.set(title = "heart diseases and cholesterol details for patients over age 60",
           ylabel = "chol")
ax0.legend(*scatter.legend_elements(), title = "target")
ax0.axhline(over_60["chol"].mean(), linestyle = "--")


ax1.scatter(x = over_60["age"],
           y = over_60["thalach"],
           c = over_60["target"])
ax1.set(title = "heart diseases and max heart rate details for patients over age 60",
          xlabel = "age", ylabel = "chol")
ax1.legend(*scatter.legend_elements(), title = "target")
ax1.axhline(over_60["thalach"].mean(), linestyle = "--")

fig.suptitle("Heart disease analysis on the basis of cholesterol and heart rate of patients over 60 years", fontsize = 14, fontweight = "bold");
```

## customizing mtaplotlib plots

```python
plt.style.available
```

```python
plt.style.use("seaborn")

fig, ax = plt.subplots(figsize = (10, 6))
ax.scatter(x = over_60["age"],
           y = over_60["chol"],
           c = over_60["target"],
           cmap = "winter")
ax.set(title = "heart-functionality and cholesterol rate of patients over age 60", xlabel = "age", ylabel = "cholesterol")
ax.legend(*scatter.legend_elements(), title = "target")
ax.axhline(over_60["chol"].mean(), linestyle = "--");
```

```python
plt.style.use("seaborn-dark")

fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1,  sharex = True, figsize = (10,10) )


ax0.scatter(x = over_60["age"],
              y = over_60["chol"],
           c = over_60["target"],
           cmap = "summer")
ax0.set(title = "heart diseases and cholesterol details for patients over age 60",
           ylabel = "chol")
ax0.set_xlim([60, 80])
ax0.set_ylim([100, 500])
ax0.legend(*scatter.legend_elements(), title = "target")
ax0.axhline(over_60["chol"].mean(), linestyle = "--")


ax1.scatter(x = over_60["age"],
           y = over_60["thalach"],
           c = over_60["target"],
            cmap = "summer")
ax1.set(title = "heart diseases and max heart rate details for patients over age 60",
          xlabel = "age", ylabel = "chol")
ax1.legend(*scatter.legend_elements(), title = "target")
ax1.axhline(over_60["thalach"].mean(), linestyle = "--")

fig.suptitle("Heart disease analysis on the basis of cholesterol and heart rate of patients over 60 years", fontsize = 14, fontweight = "bold");
```
